```python
from torchvision.transforms import RandomHorizontalFlip, ColorJitter, GaussianBlur
import glob
import os
import pandas as pd
from PIL import Image
import PIL

```

```python

ids = pd.read_csv("annotations/recognition/ids.csv", header=None)

for i in range(250,1000):
    im = Image.open(ids[0][i]).convert("RGBA")
    c = ids[1][i]
    print(ids[0][i], im.mode)

    if not os.path.exists(f"data/train/{c}"):
        os.makedirs(f"data/train/{c}")
    
    name = ids[0][i].split("/")[1].strip()
    im.save(f"data/train/{c}/0_{name}", "PNG")

    r1 = RandomHorizontalFlip(1.0)
    i1 = r1(im)
    i1.save(f"data/train/{c}/1_{name}", "PNG")

    r2 = ColorJitter(0.5, 0.5, 0.5, 0.5)
    i2 = r2(im)
    i2.save(f"data/train/{c}/2_{name}", "PNG")

    r3 = GaussianBlur(9)
    i3 = r3(im)
    i3.save(f"data/train/{c}/3_{name}", "PNG")

    
```
